## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

# Define a function to calculate the subtotal
def calc_subtotal(unit_price, qty):
    return unit_price * qty

# Apply the function to the dataframe
df['subtotal'] = df.apply(lambda x: calc_subtotal(x['unit_price'], x['qty']), axis=1)

# View the dataframe with unit_price, qty, and subtotal
df[['unit_price', 'qty', 'subtotal']].head()



```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# Define a function to calculate the shipping price
def calc_shipping(subtotal, weight):
    if weight > 50:
        return 7 * weight
    else:
        return 10 * weight
    
# Apply the function to the dataframe
df['shipping_price'] = df.apply(lambda x: calc_shipping(x['subtotal'], x['unit_weight']), axis=1)

# # View the dataframe with unit_price, unit_weight, qty, subtotal, and shipping_price
df[['unit_price', 'unit_weight', 'qty', 'subtotal', 'shipping_price']].head()

df[['unit_price', 'qty', 'subtotal', 'shipping_price']].head()



```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Define a function to calculate the total price using subtotal, shipping price, and sales tax
def calc_total(subtotal, shipping_price):
    return subtotal + shipping_price + (subtotal * 0.0925)

# Apply the function to the dataframe
df['total_price'] = df.apply(lambda x: calc_total(x['subtotal'], x['shipping_price']), axis=1)

# View the dataframe with subtotal, shipping_price, and total_price
df[['subtotal', 'shipping_price', 'total_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

# Define a function to calculate the cost
def calc_cost(unit_cost, qty, shipping_price):
    return unit_cost * qty + shipping_price

# Apply the function to the dataframe
df['cost'] = df.apply(lambda x: calc_cost(x['unit_cost'], x['qty'], x['shipping_price']), axis=1)

# View the dataframe with unit_cost, qty, shipping_price, and cost
# df[['unit_cost', 'qty', 'shipping_price', 'cost']].head(3)
df.head(3)



```

```python
# Create a column for the profit of
# each line using line cost and line price

# Define a function to calculate the profit
def calc_profit(cost, price):
    return price - cost

# Apply the function to the dataframe
df['profit'] = df.apply(lambda x: calc_profit(x['cost'], x['total_price']), axis=1)

# View the dataframe with cost, total_price, and profit
df[['cost', 'total_price', 'profit']].head(3)

df.head(3)



```
